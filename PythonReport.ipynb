#### Now to generate the report Task 2(a), i will need to use JOIN clause.

```python
# To avoid calling globals and locals everytime we can to run queries
def pysqldf(query):
    return sqldf(query, globals())
```

### The below query generates the quantities and total costs of raw materials required to produce the target sales of products. This report will be used to aggregate cost of the products

```python
# r reference raw_material, b referencing bill_of_materials, g referencing budgets tables respectively
query_1 = ''' SELECT r.Material_Name, SUM(b.Quantity_per_unit_product*g.Budget_Sales) as "Quantities", 
                SUM(b.Quantity_per_unit_product*g.Budget_Sales*Cost_per_unit) as "Total costs of raw material"
            FROM raw_materials_df r
            LEFT JOIN bill_of_materials_df b ON (b.Material_ID = r.Material_ID)
            LEFT JOIN budget_df g ON (g.product = b.Product_ID)
            GROUP BY r.Material_Name '''
pysqldf(query_1)
```

```python
# Storing in variable cost of materials to write into excel file
cost_of_Materials = pysqldf(query_1)
```

```python
cost_of_Materials
```

```python
# Writing our report to excel file named Financial Report in current working directory
cost_of_Materials.to_excel("Financial Report.xlsx", sheet_name='Cost of Raw Materials')
```

### To be able to calculate the price of each product given 30% profit, there is need to generate a report of the individual cost of raw materials in each product. Again i decide the use SQL query to generate this report. See below SQL query

```python
# r reference raw_material, b referencing bill_of_materials, g referencing budgets tables respectively
query_2 = ''' SELECT b.Product_ID,r.Material_Name, SUM(b.Quantity_per_unit_product*r.Cost_per_unit) as "Cost per Product"
            FROM bill_of_materials_df b
            LEFT JOIN raw_materials_df r ON (r.Material_ID = b.Material_ID)
            LEFT JOIN products_df p ON (p.Product_ID = b.Product_ID)
            GROUP BY r.Material_Name ORDER BY 1'''
pysqldf(query_2)
```

```python
# Storing in variable cost of material per product to write into excel file
cost_of_Material_per_Product = pysqldf(query_2)
```

```python
# Writing our report to excel file named Financial Report in current working directory
cost_of_Material_per_Product.to_excel("Report.xlsx", sheet_name='Cost of Material per product')
```
